```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
df2 = pd.read_csv(io.BytesIO(uploaded['CAISO_2021_Daily_CO2_Intensity.csv']))
```

```python
spring = df2[59:152].mean().iloc[1:]
summer = df2[152:243].mean().iloc[1:]
autumn = df2[243:334].mean().iloc[1:]
winter = pd.concat([df2[334:365],df2[0:59]]).mean().iloc[1:]

spring.plot(label='Spring')
summer.plot(label='Summer')
autumn.plot(label='Autumn')
winter.plot(label='Winter')

plt.legend()
plt.title('Seasonal Carbon Intensity Curve in 2021')
plt.xlabel("Time (24 hours)")
plt.ylabel("Carbon Intensity (mTCO2/MWh)")
```
